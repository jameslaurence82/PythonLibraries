# Indexing

```python
import pandas as pd
```

```python
oo = pd.read_csv('../../data/olympics.csv',skiprows=4)
oo.head()
```

## Index

```python
# look at type for the index
type(oo.index)
```

```python
# looking at the index for the 100th element
oo.index[100]
# returns the 100th element in that dataframe
```

```python
# oo.index[100] = 5
# TypeError: Index does not support mutable operations
```

## set_index()

```python
oo.head()
```

```python
# set_index assigns a column(series) to be the index for the dataframe
oo.set_index('Athlete')
# the first column of 0,1,2,3,4 index created with the dataframe is now using the Athlete as the index
```

```python
# set_index assigns a column(series) to be the index for the dataframe
# will revert when displaying dataframe unless ',inplace = True)' is included means that the dataframe keeps the modified layout
oo.set_index('Athlete', inplace = True)
```

```python
# dataframe now has the index set as the Athlete when using head() function
oo.head()
```

```python
# resetting the index to be the 0,1,2,3,4 and inplace means the dataframe is modified to keep this index
oo.reset_index(inplace=True)
```

```python
ath = oo.set_index('Athlete')
ath.head()
```

```python
oo.head()
```

## reset_index()

```python
# viewing the new indexed dataframe
ath.head()
```

```python
# resetting the ath dataframe index to be the 0,1,2,3,4 column
ath.reset_index(inplace=True)
ath.head()
```

## sort_index()

```python
# sort_index()
# advantage of this is that when you have a particularly large data set,
# sorting the index reduces the time to access any subset of that data. 
# You can sort objects by a label along the axis. 
ath.head()
```

```python
# setting the name as the index
ath.set_index('Athlete', inplace = True)
```

```python
# displaying head of new indexed dataframe
ath.head()
```
