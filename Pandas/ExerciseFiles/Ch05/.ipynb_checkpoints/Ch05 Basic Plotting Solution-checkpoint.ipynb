# Basic plotting challenge

```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt # if plotting only need the .pyplot function from matplotlib
# ensures plots are done inline of the code
%matplotlib inline 
```

```python
oo = pd.read_csv('../../data/olympics.csv',skiprows=4)
oo.head()
```

```python
china = oo[(oo.NOC == 'CHN') & (oo.Edition == 2008)]
china
```

### Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using:
- matplotlib
- Seaborn

```python
# Plot the number of medals achieved by the men and women of the Chinese team in Beijing 2008 
# matplotlib - bar chart
china.Gender.value_counts().plot(kind='bar');
```

```python
# Plot the number of medals achieved by the men and women of the Chinese team in Beijing 2008 
# matplotlib - pie chart
china.Gender.value_counts().plot(kind='pie');
```

```python
# Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 
# seaborn
sns.countplot(x='Gender', data = china);
```

```python
# Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 
# seaborn with hue as gender
sns.countplot(x='Medal', data = china, hue = 'Gender');
```

### Plot the number of Gold, Silver and Bronze for each gender. 

```python
# Plot the number of medals achieved by the each gender of the Chinese team in Beijing 2008 
# seaborn with palette
sns.countplot(x = 'Medal', hue = 'Gender', data = china, palette='Set1');
```

```python
# make the plot more meaningful because of the color variation between men and women
sns.countplot(x = 'Medal', hue = 'Gender', data = china, palette='bwr');
```

### Plot the number of Gold, Silver and Bronze for each gender taking into account colormaps. Address any problems with the plot.

```python
# to change the order of the x-axis use the order function
sns.countplot(x = 'Medal', hue = 'Gender', data = china, palette='bwr', order = ['Gold', 'Silver', 'Bronze']);
```
