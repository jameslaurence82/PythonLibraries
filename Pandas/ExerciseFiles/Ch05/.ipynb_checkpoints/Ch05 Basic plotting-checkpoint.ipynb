# Basic plotting

```python
import pandas as pd
```

```python
oo = pd.read_csv('../../data/olympics.csv',skiprows=4)
oo.head()
```

```python
# import the matplotlib library to do plotting
import matplotlib.pyplot as plt
```

```python
# this line of code allows the plots to occur after the code
# in the notebook
%matplotlib inline 
```

## Plot types

### What were the different sports in the first olympics? Plot them using different graphs.

```python
# sort by the Edition to get the first year the olympics occurred
fo = oo[oo.Edition == 1896]
fo.head()
```

```python
# count the values of the different Sports for the first olympics
fo.Sport.value_counts()
```

```python
# default plot created
fo.Sport.value_counts().plot()
# by adding a semi-colon at the end of the code it removes
# the matplot axis header
```

```python
# add semi-colon at the end of the code to axes header
fo.Sport.value_counts().plot();
```

### Line plot

```python
# .plot(kind='<type of plot>) <<-- line, bar, barh, pie
fo.Sport.value_counts().plot(kind='line');
```

### Bar plot

```python
# change the plot kind to bar plot
fo.Sport.value_counts().plot(kind='bar');
# this one is better for the data
```

### Horizontal bar plot

```python
# change the plot kind to barh plot
fo.Sport.value_counts().plot(kind='barh');
```

### Pie chart

```python
# change the plot kind to pie plot
fo.Sport.value_counts().plot(kind='pie');
```

## Plot colors

```python
# default line plot with the line in plum color
fo.Sport.value_counts().plot(color='plum')
```

```python
# change the bar colors to be maroon
fo.Sport.value_counts().plot(kind='bar',color='maroon');
```

## figsize()

```python
# figure size is a tuple (x,y)
# with the original line plot, the x axis titles were squeezed
# together. using figsize=(x,y) you can specify the size

fo.Sport.value_counts().plot(figsize=(3,3)); # small width, small height
```

```python
fo.Sport.value_counts().plot(figsize=(3,10)); # Small width, High Height
```

```python
fo.Sport.value_counts().plot(figsize=(10,3)); # high width, small Height
```

## Colormaps

```python
# three types of color maps
# Sequential, Diverging, Qualitative 
# The sequential should be used for representing information that
# has ordering. There is a change in lightness, often over a single 
# hue. Diverging is to be used when the information being plotted 
# deviates around a middle value. Here there are often two different 
# colors being used. And finally, the qualitative class is used to 
# represent information which does not have any ordering or relationship, 
# and is often miscellaneous colors. 
fo.Sport.value_counts().plot(kind='pie',colormap='Paired');
```

```python
fo.Sport.value_counts().plot(kind='pie',colormap='Pastel1');
```
