## Overview and validation of data
### Shape

```python
# The shape attribute returns a tuple, that's rows and columns 
# representing the dimensionality of the DataFrame. shape
oo.shape
```

```python
# extract the shape information Axis 0
oo.shape[0]
# gives the number of rows
```

```python
# extract the shape information Axis 1
oo.shape[1]
# gives the number of columns
```

### head() and tail()

```python
# to get the first 5 rows use .head() as default
oo.head()
```

```python
# to get the last 5 rows use .tail() as default
oo.tail()
```

```python
# .head(<number of rows wanted>) can be used to specify number of rows displayed
oo.head(3) # returns the first 3 rows
```

```python
# .tail(<number of rows wanted>) can be used to specify number of rows displayed
oo.tail(3) # returns the last 3 rows
```
