### List only the NOC column using both the ['..'] and dot notation. What type is this object?

```python
# list the NOC column using dot notation for series layout
# doesn't work if there is spaces in the column name
oo.NOC
```

```python
# list the NOC column using ['<column name>'] notation for series layout
# use this method if there are spaces in the column name
oo['NOC']
```

```python
# list the NOC column using [['<column name>']] notation for dataframe layout
# use this method if there are spaces in the column name
oo[['NOC']]
```

```python
# what type of object is dot notation method returning?
print(type(oo.NOC))
```

```python
# what type of object is ['<column name>'] notation method returning?
print(type(oo['NOC']))
```

```python
# what type of object is [[<column name>']] notation method returning?
print(type(oo[['NOC']]))
```

### List the Edition, City, Athlete Name and Medal columns. What type is this object?

```python
# list the types for the columns edition, city, athlete, medal columns
oo[['Edition','City', 'Athlete','Medal']]
```

```python
# what is the object type 
print(type(oo[['Edition','City', 'Athlete','Medal']]))
```
